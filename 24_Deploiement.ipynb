## 🚀 Déploiement : partager votre code Python avec le monde 🌍🐍

Vous avez écrit un super code Python. Parfait. L'étape suivante est essentielle : le **déploiement**. Déployer, c'est empaqueter et livrer votre code pour que d'autres personnes (ou d'autres systèmes) puissent l'utiliser. La bonne approche dépend de *qui* va utiliser votre code et *comment*.

Voici les méthodes les plus courantes et les plus pratiques pour déployer du code Python.

## 🖥️ Interfaces en ligne de commande (CLI)

Utilisez des bibliothèques comme `argparse` (intégrée) ou `Click` pour transformer votre code en outil en ligne de commande.

**Avantages**

* Facile à construire et à maintenir
* Très adapté à l'automatisation et aux scripts
* Familier pour les développeurs et utilisateurs avancés

**Inconvénients**

* Peu convivial pour les utilisateurs non techniques
* Nécessite l'installation de Python (et des dépendances)

**Idéal pour :** outils développeur, utilitaires internes, scripts d'automatisation

Exemple avec un outil en ligne de commande à plusieurs arguments :

```python
import argparse
def main(arg1, arg2):
    print(f"Argument 1: {arg1}, Argument 2: {arg2}")
if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="A simple CLI tool")
    parser.add_argument("arg1", type=str, help="First argument")
    parser.add_argument("arg2", type=str, help="Second argument")
    args = parser.parse_args()
    main(args.arg1, args.arg2)
```
Puis vous pouvez l'exécuter en ligne de commande comme ceci :

```bash
python my_tool.py value1 value2
```

## 🪟 Applications de bureau

Utilisez des frameworks graphiques comme `Tkinter` et/ou `PyQt` pour créer des applications avec interface graphique.

**Avantages**

* Interface visuelle conviviale
* Pas besoin de ligne de commande
* Peut être empaquetée en exécutable autonome

**Inconvénients**

* Plus complexe à développer et à maintenir
* Problèmes spécifiques aux plateformes (Windows / macOS / Linux)
* La taille finale de l'application peut être importante

**Idéal pour :** outils destinés à des utilisateurs non techniques

## 🌐 Applications web

Utilisez des frameworks comme `streamlit`, `Gradio`, `Flask` ou `Django` pour créer des applications accessibles via un navigateur.

**Avantages**

* Aucune installation nécessaire côté utilisateur
* Accessible depuis n'importe où
* Mises à jour centralisées faciles

**Inconvénients**

* Nécessite un hébergement et de la gestion serveur
* Demande des connaissances en développement web

**Idéal pour :** tableaux de bord, outils multi-utilisateurs, applications publiques

<img src="files/streamlit-nyc-app.png" alt="Streamlit NYC App" width="600" source="docs.snowflake.com"/>

## 🔌 APIs (interfaces de programmation)

Exposez vos fonctionnalités de manière programmatique avec `FastAPI`, `Flask-RESTful` ou `Django REST Framework`.

**Avantages**

* Accès indépendant du langage
* Intégration facile avec d'autres systèmes
* Bonne montée en charge

**Inconvénients**

* Pas directement utilisable par des non-développeurs
* Nécessite design et documentation d'API

**Idéal pour :** microservices, modèles ML, systèmes backend

<img src="files/fast-api.png" alt="FastAPI" width="600" source="fastapi.tiangolo.com"/>

## ☁️ Déploiement cloud et serverless

Déployez sur des plateformes comme **AWS Lambda**, **Google Cloud Functions** ou **Azure Functions**.

**Avantages**

* Pas de serveur à gérer
* Scalabilité automatique
* Paiement à l'usage

**Inconvénients**

* Latence de démarrage à froid
* Limites de temps d'exécution et de ressources

**Idéal pour :** tâches événementielles, APIs légères, jobs en arrière-plan

<img src="files/aws-lambda-function.jpg" alt="AWS Lambda" width="600" source="vticloud.io"/>

## 🧰 Distribution de packages (PyPI)

Empaquetez votre code avec `setuptools` ou `poetry` puis publiez-le sur PyPI.

**Avantages**

* Installation simple via `pip`
* Favorise la réutilisation et le versioning
* Excellent pour les bibliothèques

**Inconvénients**

* Moins adapté aux applications avec de nombreuses dépendances externes
* Demande maintenance et documentation

**Idéal pour :** bibliothèques et frameworks réutilisables

<img src="files/PyPI_logo.svg" alt="PyPI" width="600" source="pypi.org"/>

## 📦 Conteneurisation (Docker)

Utilisez Docker pour regrouper votre code, vos dépendances et l'environnement d'exécution dans un conteneur. (Vous pouvez aussi utiliser `podman` sous Linux.)

**Avantages**

* Même fonctionnement partout
* Déploiement et mise à l'échelle facilités
* Standard de l'industrie

**Inconvénients**

* Courbe d'apprentissage
* Parfois excessif pour de très petits projets

**Idéal pour :** systèmes en production, projets d'équipe, déploiement cloud
