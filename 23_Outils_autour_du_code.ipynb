# 🛠️ Outils autour du code

## 📦 Gestionnaires de paquets

Python est génial, mais la gestion des paquets peut parfois être un vrai casse-tête.

### 📦 Qu’est-ce qu’un paquet ?

Un paquet est un ensemble de modules qui peuvent être installés et utilisés dans vos projets en Python. Ce sont du code écrits par d'autres personnes.

### 🔧 Qu’est-ce qu’un gestionnaire de paquets ?

Un gestionnaire de paquets (ou package manager) est un outil qui vous aide à installer, mettre à jour et gérer des paquets dans votre environnement Python. Il simplifie la recherche et l’installation des paquets, ainsi que la gestion des dépendances entre eux.

### 🤷 Quelles sont les options ?

Il existe de nombreuses options pour gérer les paquets en Python. Les plus populaires sont :

- `pip` : le gestionnaire de paquets par défaut de Python, inclus avec la plupart des installations Python.
- `conda` : un système de gestion de paquets et d’environnements capable de gérer des paquets Python et d’autres langages.
- `uv` : un gestionnaire de paquets moderne et très rapide, écrit en Rust, qui se concentre sur la performance et la simplicité.
- `pixi` : un gestionnaire de paquets minimaliste et rapide, conçu pour être simple à utiliser et à intégrer dans les projets Python.
- `pdm` : un gestionnaire moderne de paquets et de dépendances Python visant à simplifier la gestion des projets.
- `poetry` : un outil de gestion des dépendances et de packaging qui facilite la gestion des dépendances et la création de paquets.
- `pipenv` : un outil qui combine `pip` et `virtualenv` pour offrir une gestion unifiée des dépendances et des environnements virtuels.
- `setuptools` : un outil de développement et de distribution de paquets qui inclut des fonctionnalités pour installer et gérer des paquets.

### 🐍 `pip`

`pip`, qui signifie « Pip Installs Packages » (un acronyme récursif), est le gestionnaire de paquets le plus utilisé en Python. Il permet d’installer des paquets depuis le Python Package Index (PyPI), le dépôt officiel de logiciels tiers pour Python, ainsi que depuis d’autres dépôts.

Il gère automatiquement les dépendances, en s’assurant que tous les paquets requis sont installés avec celui que vous souhaitez.

Tous les paquets installés via `pip` sont stockés dans un répertoire appelé `site-packages`, situé dans votre installation Python ou votre environnement virtuel.

Les commandes de base de `pip` incluent :

- `pip install package_name` : installe un paquet. Si aucune version n’est spécifiée, la dernière version est installée.

Pour être sûr de l’environnement Python utilisé par `pip`, vous pouvez exécuter :

```bash
python -m pip install package_name
```

L’option `-m` indique à Python d’exécuter le module `pip` comme un script, ce qui garantit que le paquet est installé dans le bon environnement.

* `pip uninstall package_name` : désinstalle un paquet.

* `pip list` : affiche la liste des paquets installés.

* `pip freeze` : affiche la liste des paquets installés dans un format permettant de recréer l’environnement. Cependant, sauf si vous souhaitez recréer exactement le même environnement pour la production, préférez `pip list`.

* `pip show package_name` : affiche des informations sur un paquet spécifique.

* `pip install -r requirements.txt` : installe les paquets listés dans un fichier de dépendances.

Ne mettez pas toutes vos dépendances dans un seul fichier sans précaution. Vous devriez spécifier les versions à l’aide d’opérateurs de comparaison afin d’éviter les problèmes de compatibilité. Par exemple :

```
pandas >=2.0.0
numpy >=1.21.0
requests <=2.25.0
```

Vous pouvez aussi utiliser `pip-compile` de `pip-tools` pour générer un fichier `requirements.txt` avec des versions figées à partir d’un fichier `requirements.in`. Par exemple, si votre fichier `requirements.in` ressemble à ceci :

```
# requirements.in
pandas>=2.1
requests>=2.31
```

Vous pouvez alors exécuter :

```bash
pip-compile requirements.in
```

Cela générera un fichier `requirements.txt` contenant les versions exactes des paquets et de leurs dépendances :

```# requirements.txt
pandas==2.1.0
numpy==1.24.3
python-dateutil==2.8.2
pytz==2023.3
requests==2.31.0
certifi==2023.5.7
charset-normalizer==3.1.0
idna==3.4
urllib3==1.26.15
```

* `pip install --upgrade package_name` : met à jour un paquet installé vers la dernière version.

* `pip check` : vérifie les dépendances cassées parmi les paquets installés.

* `pip install -e .` : installe un paquet en mode éditable, ce qui permet de modifier le code source et de voir immédiatement les changements reflétés dans le paquet installé. En effet, avec `pip install package_name`, les fichiers sont copiés dans le répertoire `site-packages`. Les modifications du code source original ne sont donc pas prises en compte sans réinstallation.

### 🐍 Conda

`conda` est un autre gestionnaire de paquets très populaire, capable de gérer des paquets pour Python et d’autres langages. Il est souvent utilisé en data science et en calcul scientifique, car il gère bien les dépendances complexes et fournit des binaires précompilés pour de nombreux paquets.

Pourquoi utiliser conda plutôt que pip ?

* Gestion des environnements : `conda` intègre nativement la création et la gestion d’environnements virtuels, ce qui permet d’isoler facilement les dépendances entre projets. Il remplace donc à la fois `pip` et `virtualenv`.

* Support multi-langages : `conda` peut gérer des paquets pour plusieurs langages de programmation, pas seulement Python. C’est un gros avantage si vous travaillez avec plusieurs langages.

* Binaires précompilés : `conda` fournit des binaires précompilés pour de nombreux paquets, ce qui fait gagner du temps et évite des compilations complexes.

* Meilleure résolution des dépendances : `conda` utilise un algorithme de résolution des dépendances plus sophistiqué que `pip`, ce qui permet de réduire les conflits entre paquets.

Si vous travaillez dans un environnement conda, privilégiez `conda` pour installer les paquets lorsque c’est possible. Si un paquet n’est pas disponible via `conda`, vous pouvez utiliser `pip` en solution de secours.

* `conda install package_name` : installe un paquet.
* `conda remove package_name` : désinstalle un paquet.
* `conda list` : liste les paquets installés.
* `conda update package_name` : met à jour un paquet installé vers la dernière version.

Bonne pratique : de temps en temps, exécutez `conda clean --all` pour supprimer les paquets et caches inutilisés. Cela peut libérer plusieurs Go d’espace disque et garder votre environnement conda propre.

### ⚡ uv

`uv` est un gestionnaire de paquets moderne et très rapide, écrit en Rust, qui met l’accent sur la performance et la simplicité. Il vise à remplacer les autres gestionnaires de paquets et outils d’environnements virtuels en proposant une solution unifiée et efficace.

Les commandes de base de `uv` pour l’installation de paquets ressemblent beaucoup à celles de `pip` et `conda` :

* `uv install package_name` : installe un paquet.
* `uv remove package_name` : désinstalle un paquet.
* `uv list` : liste les paquets installés.
* `uv update package_name` : met à jour un paquet installé vers la dernière version.

**CEPENDANT**, sauf si vous voulez simplement tester quelque chose, ce n’est pas la bonne manière d’utiliser `uv`. Vous devriez utiliser `uv add package_name` pour ajouter un paquet à votre projet. Cela créera un fichier `uv.lock` pour gérer vos dépendances. Ensuite, vous pouvez utiliser `uv sync` pour installer toutes les dépendances du projet.

Nous en reparlerons dans la section suivante sur les environnements virtuels.

Le seul inconvénient est que `uv` est relativement récent et peut ne pas proposer autant de paquets que `pip` ou `conda`.

### 🧪 Pixi

`pixi` est une sorte de `conda 2.0` qui utilise `uv` en interne pour certaines tâches. Il est un peu plus complexe et un peu plus lent que `uv`, mais il offre une interface plus riche et des fonctionnalités supplémentaires pour la gestion de projets non seulement Python mais dans d'autres langages également.


### 📜 Poetry

Poetry est un outil de gestion des dépendances et de packaging pour Python qui simplifie la gestion des dépendances et la création de paquets. Il offre une interface simple et intuitive pour gérer les dépendances, créer des environnements virtuels et construire des paquets.

Vous pouvez commencer avec Poetry en créant un nouveau projet :

```bash
poetry init
```

Cette commande vous guidera dans la création d’un fichier `pyproject.toml`, utilisé pour gérer les dépendances et la configuration du projet.

Pour installer des paquets, vous pouvez utiliser les commandes suivantes :

* `poetry add package_name` : installe un paquet et l’ajoute au fichier `pyproject.toml`.
* `poetry remove package_name` : désinstalle un paquet et le supprime du fichier `pyproject.toml`.
* `poetry show` : liste les paquets installés.
* `poetry update package_name` : met à jour un paquet installé vers la dernière version.

Pour recréer l’environnement sur une autre machine, il suffit ensuite d’exécuter :

```bash
poetry install
```

Cette commande lit le fichier `pyproject.toml` et installe toutes les dépendances requises dans un nouvel environnement virtuel.

### 🏗️ PDM

`PDM` (Python Development Master) est un gestionnaire moderne de paquets et de dépendances Python qui vise à simplifier la gestion des projets Python. Il utilise le fichier `pyproject.toml` pour gérer les dépendances et la configuration du projet, de manière similaire à Poetry.

Il pourrait remplacer `poetry` à l’avenir, car il est plus rapide et plus léger.

* Gestion simplifiée des dépendances : PDM facilite l’ajout, la suppression et la mise à jour des dépendances via des commandes simples.
* Gestion des environnements virtuels : PDM peut créer et gérer automatiquement des environnements virtuels pour vos projets.
* Support de la PEP 582 : PDM prend en charge la PEP 582, qui permet de gérer les dépendances sans utiliser d’environnements virtuels.
* Construction et publication de paquets : PDM fournit des outils pour construire et publier des paquets Python sur PyPI ou d’autres dépôts.

### 🧪 Pipenv

`Pipenv` est un outil qui combine `pip` et `virtualenv` pour offrir une gestion unifiée des dépendances et des environnements virtuels. Il simplifie la création et la gestion des environnements virtuels, ainsi que l’installation et la mise à jour des paquets.

Cependant, son développement a ralenti ces dernières années, et d’autres outils comme `poetry` et `pdm` ont gagné en popularité comme alternatives.

### ⚙️ Setuptools

`Setuptools` est un outil de développement et de distribution de paquets qui inclut des fonctionnalités pour installer et gérer des paquets. Aujourd’hui, il est principalement utilisé pour le packaging et la distribution de projets Python plutôt que pour la gestion des dépendances.

### 📝 TL;DR

* Utiliser `uv` pour la gestion des paquets **et** des environnements virtuels est la meilleure option dans la plupart des cas.
* Si vous utilisez des dépendances complexes, travaillez ne data science ou utilisez plusieurs langages, essayez `pixi`.
* Si cela ne marche toujours pas, utilisez `conda`.
* `pip` reste suffisant dans de nombreux cas.
* Envisagez `pdm` ou `poetry` pour une gestion moderne des dépendances et du packaging, surtout si vous souhaitez des projets reproductibles et facilement partageables.


# 🌐 Environnements virtuels

Un problème courant en Python : votre projet nécessite Python 3.11, mais votre système a Python 3.9. Ou vous travaillez sur plusieurs projets qui demandent des versions différentes.

Vous pourriez alors installer des versions différentes de Python sur votre système, mais cela peut rapidement devenir un cauchemar à gérer : c’est là que les environnements virtuels entrent en jeu.

Un environement virtuel n'est pas une machine virtuelle, mais plutôt un environnement isolé pour votre projet Python. Il vous permet d'avoir différentes versions de Python et de paquets installés pour différents projets, sans interférer les uns avec les autres.

Concrètement ce que font la plupart des outils d'environnements virtuels, c'est de créer un dossier dans lequel ils installent une copie de l'interpréteur Python et un répertoire `site-packages` pour les paquets.

Ensuite, ils configurent les variables d'environnement pour que lorsque vous activez cet environnement virtuel, votre terminal utilise cette version de Python et les paquets installés dans ce dossier.

Cela permet d'éviter les conflits entre les paquets et garantit que chaque projet dispose de son propre environnement isolé.

Il existe de nombreuses options pour créer des environnements virtuels en Python, notamment :

- `conda` : Un système de gestion de paquets et d'environnements qui peut créer des environnements virtuels pour Python mais aussi d'autres langages.

- `uv` : Un outil rapide et moderne pour créer et gérer des environnements virtuels en Python.
  
- `pyenv` : Un outil populaire pour gérer plusieurs versions de Python et créer des environnements virtuels.
  
- `venv` : Un module intégré à Python 3 pour créer des environnements virtuels légers.

- `virtualenv` : Un outil tiers populaire pour créer des environnements virtuels.


## 🐍 Conda

`conda` est un choix très populaire pour la gestion des environnements virtuels, en particulier en data science et en calcul scientifique. Vous pouvez créer un nouvel environnement conda avec la commande suivante (supposons que vous voulez Python 3.11 et que vous voulez appeler votre environnement "myenv") :

```bash
conda create --name myenv python=3.11
```

Ensuite, vous pouvez activer l'environnement avec :

```bash
conda activate myenv
```

Pour désactiver l'environnement, exécutez simplement :

```bash
conda deactivate
```

Les bonnes pratiques sont :

- créer un nouvel environnement conda pour chaque projet afin d'éviter les conflits de dépendances.
  
- Essayez de ne rien installer dans l'environnement de base.
  
- utiliser `conda env export > environment.yml` pour exporter la configuration de l'environnement dans un fichier YAML, qui peut être utilisé pour recréer l'environnement sur une autre machine avec `conda env create -f environment.yml`.

Ce fichier yaml pourrait ressembler à ceci :

```yaml
name: myenv
channels:
  - defaults
dependencies:
  - python=3.11
  - numpy=1.24.3
  - pandas=2.1.0
  - pip:
    - requests==2.31.0
```

Cependant, comme conda est un outil pensé pour utiliser différents langages, il ne lit pas vraiment les fichiers de dépendances Python standards comme `requirements.txt` ou `pyproject.toml`. Il existe des moyens de contourner cela en intégrant `pip` dans votre gestion d'environnement conda, mais ce n'est pas idéal.

## ⚡ UV

`uv`, comme vu précédemment, est un outil qui gère à la fois les paquets et les environnements virtuels. C’est une excellente option pour la plupart des projets Python, car il est rapide, moderne et facile à utiliser.

### Installation

La meilleure façon d’installer `uv` est d’utiliser le [script d’installation officiel](https://docs.astral.sh/uv/getting-started/installation/#installation-methods).

Il est également possible de l'installer avec `pip`, mais ce n’est pas recommandé car cela peut causer des problèmes de compatibilité. Évitez également de l’installer dans un environnement `conda` car cela peut aussi causer des problèmes de compatibilité et des bugs assez subtils.

Uv est un outil de haut niveau qui gère lui-même son environnement d’exécution, donc il est préférable de l’installer globalement sur votre système.

### 🏗️ Créer et structurer un projet

Maintenant que `uv` est installé, créons un vrai projet Python. `uv` génère une structure de fichiers propre et moderne.

#### ✨ Créer un projet

Placez-vous dans le dossier où vous rangez vos projets de code et lancez :

```bash
uv init my_project
cd my_project
```

Ou bien, si vous venez de cloner un repo git existant qui est vide ou quasi-vide:

```bash
git clone <url-du-repo>
cd my_project
uv init
```

`uv` crée automatiquement cette structure :

```
my_project/
├── pyproject.toml    # Configuration du projet et dépendances
├── README.md         # Documentation
├── .python-version   # Version Python à utiliser
└── hello.py          # Script d'exemple
```

Si certains fichiers existent déjà, `uv` les laisse intacts. Par exemple, si vous avez déjà un `README.md`, `uv` ne le remplacera pas.

#### 📄 Comprendre pyproject.toml

Ce fichier est le cœur de votre projet. son contenu devrait ressembler à ceci :

```toml
[project]
name = "my_project"
version = "0.1.0"
description = "Add your description here"
readme = "README.md"
requires-python = ">=3.11"
dependencies = []
```

Le fichier `pyproject.toml` est un standard Python (PEP 621) que comprennent pip, poetry, et uv. Vos dépendances (les librairies à installer et les versions compatibles) seront listées dans `dependencies`.

#### ➕ Ajouter une dépendance

Pour ajouter une librairie à votre projet :

```bash
uv add requests
```

Que se passe-t-il ? `uv` fait trois choses :

1.  Télécharge le paquet et ses sous-dépendances
2.  Met à jour `pyproject.toml` avec la nouvelle dépendance
3.  Génère/met à jour `uv.lock` avec les versions exactes

Vous verrez `requests` apparaître dans votre `pyproject.toml` :

```toml
[project]
dependencies = [
    "requests>=2.32.0",
]
```
#### ✅ Et voilà ! Vous pouvez désormais lancer votre programme !

Avec `uv` pas besoin d’activer manuellement un environnement virtuel. Ni d'installer manuellement les dépendances. **Tout est automatique**. Du moment que votre fichier `pyproject.toml` est à jour, `uv` s’occupe du reste!

Pour exécuter le script d’exemple :

```bash
uv run hello.py
```

Vous devriez voir apparaître un `Hello from mon-projet!` sur votre terminal. En lançant la commande `uv run hello.py`, `uv` a créé automatiquement l’environnement virtuel (vous verrez un dossier `.venv/`), téléchargé la bonne version de Python, installé toutes les dépendances et exécuté le script.

#### ⚡ Pourquoi `uv` est si rapide ?

`uv` est très rapide parce qu’il est implémenté comme un binaire natif écrit en rust, avec un résolveur de dépendances performant, des téléchargements parallélisés, un cache global efficace (il évite de retélécharger inutilement des paquets) et une logique conçue pour éviter tout travail inutile.

#### 🔄 Synchroniser l’environnement avec `uv sync`

En lançant un `uv run`, `uv` lance aussi un `uv sync` implicitement.

Cette commande garantit que votre environnement correspond exactement à ce qui est déclaré dans `uv.lock`. Vous la lancer manuelement à tout moment pour vous assurer que votre environnement est à jour (sans avoir à lancer votre script).

Concrètement, `uv sync` :

*   Crée l’environnement virtuel `.venv/` s’il n’existe pas
*   Installe les dépendances aux versions exactes du lock file
*   Supprime les paquets qui ne sont plus déclarés

C’est la commande à lancer quand vous clonez un projet ou après un `git pull` qui a modifié les dépendances.

#### 🔒 Le fichier uv.lock

Le fichier `uv.lock` contient les versions exactes de toutes les dépendances, y compris les sous-dépendances. Il garantit que tous les développeurs, la CI, et la production utilisent exactement les mêmes versions.

Il est très utile : versionnez-le avec Git! (ne l’ajoutez donc **pas** au `.gitignore`).

### ⚙️ Davantage de fonctionnalités de `uv`

#### 🪐 `uv` et les notebooks Jupyter

Si vous utilisez des notebooks Jupyter, uv peut aussi gérer les dépendances pour eux. Il suffit d’ajouter `ipykernel` et/ou `jupyterlab` à vos dépendances :

```bash
uv add ipykernel
```

- Si vous ouvrez le dossier de votre projet avec `vscode`, celui-ci devrait détecter l'environnement virtuel créé par uv et vous proposer de l'utiliser pour vos notebooks.

- Si vous préférez utiliser `jupyterlab` directement, vous pouvez aussi l’ajouter à vos dépendances. C'est très facile, lisez la [documentation officielle](https://docs.astral.sh/uv/guides/integration/jupyter/) à ce sujet.

#### 🛠️ Dépendances de développement

Certaines dépendances ne sont utiles que pendant le développement : les outils de test, les linters, les formateurs de code. Elles ne doivent pas être installées en production.

uv les distingue avec l’option `--dev` :

```bash
uv add --dev pytest ruff mypy
```

Ces dépendances apparaissent dans une section séparée et ne seront pas installées quand vous déployez en production avec `uv sync --no-dev`.

#### 🎯 Spécifier une version

Parfois vous avez besoin d’une version précise, par exemple pour éviter un bug ou assurer la compatibilité :

```bash
# Au moins la version 2.28
uv add 'requests>=2.28'

# Exactement cette version
uv add 'requests==2.28.0'

# Entre deux versions
uv add 'requests>=2.28,<3.0'
```

#### ➖ Supprimer une dépendance
Pour retirer une dépendance devenue inutile :

```bash
uv remove requests
```

uv met à jour `pyproject.toml`, `uv.lock`, et désinstalle le paquet de l’environnement.

#### 🔄 Mettre à jour les dépendances

```bash
# Regénérer le lock file
uv lock --upgrade

# Mettre à jour un paquet spécifique
uv lock --upgrade-package requests
```

#### ❓ Dépendances conditionnelles

Pour installer un paquet uniquement sur certaines plateformes :

```bash
uv add 'pywin32; sys_platform == "win32"'
uv add 'uvloop; sys_platform != "win32"'
```

#### 🏃 Exécution de commandes avec uv run

`uv run` exécute une commande dans l’environnement du projet :

```bash
# Exécuter un script
uv run python script.py

# Lancer un module
uv run python -m http.server

# Exécuter un outil installé en dépendance dev
uv run pytest
uv run ruff check .
```

#### 📜 Scripts personnalisés

Définissez des raccourcis dans `pyproject.toml` :

```toml
[tool.uv.scripts]
serve = "python -m http.server"
test = "pytest -v"
lint = "ruff check ."
format = "ruff format ."
```

Puis exécutez-les :

```bash
uv run serve
uv run test
```

#### 👁️ Voir les versions disponibles

Commencez par lister ce qui est disponible sur votre système et ce que uv peut télécharger :

```bash
uv python list
```

La sortie montre deux types d’entrées :

```
cpython-3.13.0-linux-x86_64-gnu    <download available>   # Peut être téléchargé
cpython-3.12.3-linux-x86_64-gnu    /usr/bin/python3.12    # Déjà installé
cpython-3.11.10-linux-x86_64-gnu   <download available>   # Peut être téléchargé
```

uv détecte automatiquement les versions Python déjà présentes sur votre système (installées par le gestionnaire de paquets de votre OS, par exemple). Les versions marquées `<download available>` peuvent être installées par uv.

#### ⬇️ Installer une version spécifique

Supposons que votre projet nécessite Python 3.11 et que vous ne l’avez pas. Installez-le simplement :

```bash
uv python install 3.11
```

uv télécharge une version précompilée depuis le registre Python. C’est beaucoup plus rapide que de compiler depuis les sources comme le fait pyenv.

#### 📂 Où sont stockées les versions ?

Cela dépend de votre OS, mais en uv devrait les installer dans `~/.local/share/uv/python/`. Chaque version est isolée et n’interfère pas avec le Python système.

#### 📌 Fixer la version pour un projet

Pour garantir que tous les développeurs utilisent la même version de Python, fixez-la dans le projet :

```bash
cd my_project
uv python pin 3.11
```

Cette commande crée un fichier `.python-version` contenant simplement `3.11`. Quand quelqu’un clone le projet et exécute `uv sync` ou `uv run`, uv utilise automatiquement cette version (et la télécharge si nécessaire).

C’est un fichier standard reconnu aussi par pyenv et d’autres outils, donc même si vos collègues n’utilisent pas uv, ils sauront quelle version utiliser.

## 📝 TL;DR

- En priorité, utilisez `uv` à la fois pour gérer vos dépendances et vos venvironnements virtuels. C’est la meilleure option dans la plupart des cas.

- Si vous avez besoin de dépendances complexes, utilisez beaucoup de librairies de machine learning ou utilisez d'autres langages, faites plutôt appel à `pixi`.
