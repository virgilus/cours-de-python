# 🐍 Interpréteur Python

## 🔍 Comment Python fonctionne-t-il sous le capot ?

Comprendre le fonctionnement interne de Python aide à mieux saisir ses caractéristiques de performance, le débogage, et même à écrire un meilleur code, en particulier si vous venez d'autres langages ou si vous voulez créer des applications multithreadées ou hautes performances.

## 💻 Comment un programme s'exécute-t-il sur votre ordinateur ?

Il y a essentiellement deux composants importants : le processeur (CPU, par exemple Intel, ARM, AMD) et la mémoire.

### 📜 Exécution du code source : du texte aux instructions machine

Quand vous écrivez du code source (Python, Java, C, C++), vous créez simplement un **fichier texte** stocké sur le disque.

Cependant, le **processeur ne comprend pas directement ce code source**. Un CPU n'exécute que du **code machine**, c'est-à-dire des instructions binaires de bas niveau (0 et 1). Le code source doit donc être traduit avant exécution. C'est le rôle du compilateur : traduire un code source vers du code machine pour une plateforme donnée (processeur + système d'exploitation).

Le résultat est toujours un fichier binaire contenant des instructions machine.

### ⚙️ Langages compilés : l'approche classique (compilation anticipée)

Dans les langages compilés comme **C, C++ ou Go**, cette traduction est effectuée par un **compilateur**.

Un compilateur :

* prend le **code source** en entrée,
* le traduit en **code machine** pour un processeur et un OS donnés,
* produit un **exécutable binaire** en sortie.

Ce binaire contient des instructions que le processeur peut exécuter directement.

Quand vous double-cliquez sur ce binaire, le système d'exploitation :

1. charge le binaire en mémoire,
2. demande au processeur d'exécuter ses instructions.

C'est le flux d'exécution standard des langages compilés.

## 🐍 Python : un modèle d'exécution différent

Python fonctionne différemment.

### 🔧 L'interpréteur Python

Supposons un fichier `hello.py` :

```python
print("Hello, world")
```

Pour l'exécuter :

```bash
python hello.py
```

Point clé :

> **Vous n'exécutez pas directement le fichier Python. Vous exécutez l'interpréteur Python.**

La commande `python` lance l'**interpréteur Python**, qui est lui-même un **programme binaire compilé**. Le code machine chargé en mémoire est celui de l'interpréteur, pas celui de `hello.py`.

### 🔄 Compilateur Python et machine virtuelle Python (PVM)

Conceptuellement, l'interpréteur Python comprend deux composants principaux :

1. **Le compilateur**
2. **La machine virtuelle Python (PVM)**

Voyons comment ils fonctionnent ensemble.

#### 📝 Étape 1 : compilation en bytecode

Quand vous lancez :

```bash
python3 hello.py
```

le nom de fichier `hello.py` est transmis à l'interpréteur.

Le composant **compilateur** :

* lit le code source Python,
* le traduit en une **représentation intermédiaire** appelée **bytecode**.

Ce bytecode est :

* un format binaire (0 et 1),
* **pas du code machine**,
* **pas spécifique à un processeur**,
* conçu pour être exécuté par la machine virtuelle Python.

---

#### ▶️ Étape 2 : exécution par la machine virtuelle Python

La **PVM** est le composant qui comprend le bytecode Python.

Son rôle ressemble à celui d'un CPU :

* elle lit les instructions bytecode,
* les interprète,
* exécute leur comportement via le matériel sous-jacent.

Le processeur n'exécute donc jamais directement le bytecode Python. Il exécute les instructions machine de l'interpréteur Python, qui interprète ensuite le bytecode.

C'est pourquoi Python est souvent décrit comme un **langage interprété**, même s'il y a bien une phase de compilation.

## 🔬 Inspection du bytecode

### 🛠️ Compiler sans exécuter

Pour compiler un fichier Python sans l'exécuter :

```bash
python -m py_compile hello.py
```

Cette commande :

* compile `hello.py` en bytecode,
* stocke le résultat dans le dossier `__pycache__`.

Dans ce dossier, vous verrez un fichier comme :

```text
hello.cpython-311.pyc
```

L'extension `.pyc` indique un **fichier bytecode**.

### 👀 Afficher le fichier bytecode

Si vous affichez ce fichier avec :

```bash
cat hello.cpython-311.pyc
```

(ou `type` sous Windows), vous verrez surtout des caractères illisibles. C'est normal : le bytecode est un **format binaire**, non prévu pour être lu par un humain.

Vous pouvez voir des fragments comme `"Hello, world"`, mais la majorité du contenu est binaire.

## 📖 Désassembler le bytecode pour les humains

Pour obtenir une version lisible du bytecode, Python fournit le module `dis` (*disassemble*) :

```bash
python3 -m dis hello.py
```

Cette commande affiche une suite d'**instructions bytecode**, par exemple :

```
LOAD_CONST
LOAD_NAME
CALL_FUNCTION
RETURN_VALUE
```

Ces instructions sont comprises par la PVM et exécutées étape par étape.

## 📌 Remarque sur les implémentations

### 🏆 CPython : l'implémentation de référence

Python est une **spécification de langage**, pas un unique programme. Ce qui est décrit ici correspond à **CPython**, l'implémentation de référence et la plus utilisée.

Quand vous installez Python depuis **python.org**, vous installez CPython. Dans la pratique, c'est généralement ce qui tourne sur votre machine.

D'autres langages comme **Java** suivent un modèle proche : code source → bytecode → exécution par machine virtuelle.

### 🔀 Autres implémentations

Il existe d'autres implémentations de Python, par exemple :

* **PyPy** : implémentation alternative orientée performance, avec compilation Just-In-Time (JIT).
* **Jython** : Python implémenté sur la JVM.
* **IronPython** : Python pour l'écosystème .NET.
