## Round function behavior for decimal

Python follows the "round half to even" strategy, also known as "bankers' rounding". This means that when a number is exactly halfway between two integers, it rounds to the nearest even integer. For example:

```python
print(round(2.5))  # Output: 2
print(round(3.5))  # Output: 4
print(round(4.5))  # Output: 4
print(round(5.5))  # Output: 6
```

This behavior helps to reduce rounding bias in calculations that involve many rounding operations.

It is compliant with the [IEEE 754](https://en.wikipedia.org/wiki/IEEE_754#Rounding_rules) standard for floating-point arithmetic, which is widely used in computer systems.

Short answer: **because computers use binary, not decimal**, and `0.1`, `0.2`, and `0.3` cannot be represented exactly in binary floating-point.

Now let’s unpack that properly (this is a *classic*, and it connects directly to your previous questions about IEEE 754 and rounding).

## 🔢 Decimal approximation

### A strange behavior

````python
0.1 + 0.2 == 0.3
````

### Base 10 vs base 2

In base 10 we write:

```
0.1 = 1/10
```
And it outputs a clean, finite decimal.

But computers store floats using **binary fractions**:

```
0.1₁₀ = 0.0001100110011001100110011…₂
```

That `0011` **repeats forever**. So the computer must **truncate** it. It's the same problem then:

```
1 / 3 = 0.333333...
```
in base 10.


## 🧪 What Python actually stores

Let’s peek under the hood:

```python
>>> 0.1
0.1
```

Looks fine… but that’s just **pretty printing**.

Internally:

```python
>>> format(0.1, '.17f')
'0.10000000000000001'

>>> format(0.2, '.17f')
'0.20000000000000001'

>>> format(0.3, '.17f')
'0.29999999999999999'
```

So when you do:

````python
0.1 + 0.2
````

Because :

```python
0.30000000000000004 ≠ 0.29999999999999999
0.1 + 0.2 == 0.3
```
Python outputs `False`. The **floating-point equality is exact**, not approximate.


**This is NOT a Python bug**, it’s a consequence of how floating-point numbers are represented in binary. And it happens in the most popular programming languages too.

### ✅ The correct way to compare floats

- You could try to use a tolerance:

````python
abs((0.1 + 0.2) - 0.3) < 1e-9
````

But, as usual in Python, there is a better way if we use the math library (if you're curious, you can find the original C code [here](https://github.com/python/cpython/blob/main/Modules/mathmodule.c#L2586))

- By using the `math.isclose()` function:

````python
import math
math.isclose(0.1 + 0.2, 0.3)
````

- Use the decimal module for high-precision decimal arithmetic: 

````python
from decimal import Decimal

Decimal("0.1") + Decimal("0.2") == Decimal("0.3")
````
